```python
import pandas as pd
import numpy as np
import re
import nltk
pd.set_option('display.max_columns', None)
```

```python
## Data Preprocessing
```

```python

df['clean_plot'] = df['Plot'].str.lower()  ## Converted the sentence into the lower Case String
df['clean_plot'] = df['clean_plot'].apply(lambda x: re.sub('[^a-zA-Z]', ' ', x))  ## Removing the additional symbols from the string
df['clean_plot'] = df['clean_plot'].apply(lambda x: re.sub('\s+', ' ', x))  # Removing the extra spaces.
```

```python
df['clean_plot']
```

```python
## Craeting the tokenization for the sentence
```

```python
import nltk
nltk.download('punkt')  ## NLTK requires some data files for tokenization
```

```python
df['clean_plot'].fillna('', inplace=True)  ## fillna method is used to handle missing values.

# Tokenize the 'clean_plot' column
df['clean_plot'] = df['clean_plot'].apply(lambda x: nltk.word_tokenize(x))

# Display the tokenized 'clean_plot' column
print(df['clean_plot'])
```

```python
import nltk
from nltk.corpus import stopwords
```

```python
nltk.download('stopwords')
```

```python
stop_words = set(nltk.corpus.stopwords.words('english'))

# Tokenize and filter words
plot = []
for sentence in df['clean_plot']:
    temp = []
    for word in sentence:
        if word not in stop_words and len(word) >= 3:
            temp.append(word)
    plot.append(temp)
```

```python
df['clean_plot'] = plot
```

```python
df['clean_plot']
```

```python
df['Genre'] = df['Genre'].apply(lambda x: x.split(',') if isinstance(x, str) else x)  ##takes each element (a string) in the 'Genre' column and splits it into a list of values using a comma , as the delimiter

df['Actors'] = df['Actors'].apply(lambda x: x.split(',')[:4] if isinstance(x, str) else x) ## This operation is used to limit the number of actors to the top four
df['Director'] = df['Director'].apply(lambda x: x.split(',') if isinstance(x, str) else x)
```

```python
df['Actors'][0]
```

```python
## Created a new function and performed data cleaning
```

```python
def clean(sentence):
    temp = []
    for word in sentence:
        temp.append(word.lower().replace(' ', ''))  ## Converts word to lowercase and then removed the extra spaces
    return temp
```

```python
df['Genre'] = [clean(x) for x in df['Genre']]
df['Actors'] = [clean(x) for x in df['Actors']]
df['Director'] = [clean(x) for x in df['Director']]
```

```python
df['Actors'][0]
```

```python
## created a new list l by concatenating the values from specific columns in the DataFrame
```

```python
columns = ['clean_plot', 'Genre', 'Actors', 'Director']
l = []
for i in range(len(df)):
    words = ''
    for col in columns:
        words += ' '.join(df[col][i]) + ' '
    l.append(words)
l
```

```python
df['clean_input'] = l
df = df[['Title', 'clean_input']]
df.head()
```

```python
## Feature Extraction
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer  ## TF-ID Tools for converting text data into numerical formats
tfidf = TfidfVectorizer()
features = tfidf.fit_transform(df['clean_input'])
```

```python
## Cosine Similarity measure of similarity between two non-zero vectors(rows) defined in a feature matrix
```

```python
from sklearn.metrics.pairwise import cosine_similarity
cosine_sim = cosine_similarity(features, features) ## It shows like the matrix(i,j)
print(cosine_sim)
```

```python
## Movie Recommendation
```

```python
index = pd.Series(df['Title'])
index.head()
```

```python
def recommend_movies(title):
    movies = []
    idx = index[index == title].index[0]
    # print(idx)
    score = pd.Series(cosine_sim[idx]).sort_values(ascending=False)
    top10 = list(score.iloc[1:11].index)
    # print(top10)

    for i in top10:
        movies.append(df['Title'][i])
    return movies
```

```python
recommend_movies('The Dark Knight Rises')
```
